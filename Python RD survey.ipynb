```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('/Users/mr.basix/Documents/Sync/Academics/USF/Research/Reverse Diet/RDS Clean.xlsx', engine='openpyxl')
```

```python
df_cleaned = df.drop(columns=['ResponseId', 'Q1', 'Q2'])
```

```python
#What category does a majority of your clientele (coaching) fall under? 

value_counts = df_cleaned['Q3 clean'].value_counts()


percentages = ((value_counts / value_counts.sum()) * 100).round(0)


print(percentages)


# Define a dictionary to map the old values to the new ones
rename_map = {
    1: 'Competitive physique athletes',
    2: 'Lifestyle/General population',
    3: 'Strength Athletes',
    4: 'Team sports/Athletic performance'
}

# Rename the categories in 'Q3 clean'
df_cleaned['Q3 clean'] = df_cleaned['Q3 clean'].replace(rename_map)

# Count occurrences of each unique value in 'Q3 clean' after renaming
value_counts = df_cleaned['Q3 clean'].value_counts()

# Calculate percentages
percentages = (value_counts / value_counts.sum()) * 100

# Create the pie chart
plt.figure(figsize=(8, 8))  # Set the figure size for better readability
plt.pie(percentages, labels=percentages.index, autopct='%1.0f%%', startangle=140)
plt.title('Percentage of Respondents by Q3 clean Category')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.

# Display the pie chart
plt.show()

#Bar Graph 
value_counts = df_cleaned['Q3 clean'].value_counts()

percentages = (value_counts / value_counts.sum()) * 100


plt.figure(figsize=(10, 6))  # Set the figure size for better readability
plt.bar(percentages.index, percentages, color='skyblue')
plt.title('Percentage of Respondents by Q3 clean Category')
plt.xlabel('Category')
plt.ylabel('Percentage')
plt.xticks(rotation=45)  # Rotate category labels for better readability

# Display the bar graph
plt.show()
```

```python
# Ensure the column exists
if 'Q10 clean (Male kcal)' in df_cleaned.columns:
    # Calculate the average increase
    average_increase = df_cleaned['Q10 clean (Male kcal)'].mean()
    
    # Round the results to one decimal place
    average_increase_rounded = round(average_increase, 1)
    relative_increase_rounded = round(average_increase_rounded / 100, 1)  # per kg
    percentage_increase_rounded = round((average_increase_rounded / 2000) * 100, 1)  # Convert to percentage
    
    # Print out the rounded results
    print(f"Average daily caloric increase: {average_increase_rounded} calories")
    print(f"Relative daily caloric increase per kg body weight: {relative_increase_rounded} calories")
    print(f"Percentage increase relative to 2000 calories: {percentage_increase_rounded}%")
else:
    print("Column 'Q10 clean (Male kcal)' does not exist in the DataFrame. Please check the correct column name.")
```

```python

# Check if both columns exist in the DataFrame
if 'Q10 clean (Male kcal)' in df_cleaned.columns and 'Q27 Clean (Male Macro)' in df_cleaned.columns:
    # Combine the two series, ignoring NaN values from both
    combined = pd.concat([df_cleaned['Q10 clean (Male kcal)'].dropna(), df_cleaned['Q27 Clean (Male Macro)'].dropna()])
    
    # Calculate the average of the combined data, excluding NaN values
    combined_average = combined.mean()
    
    # Calculate the rounded average (absolute value)
    absolute_average_rounded = round(combined_average, 1)
    
    # Calculate the relative value based on body weight (assuming 100kg)
    relative_average_rounded = round(absolute_average_rounded / 100, 1)  # per kg
    
    # Calculate the percentage increase based on a standard (e.g., 2000 calories)
    percentage_average_rounded = round((absolute_average_rounded / 2000) * 100, 1)  # Convert to percentage
    
    # Print out the results
    print(f"Average combined value (absolute): {absolute_average_rounded} calories")
    print(f"Relative value based on body weight (100kg): {relative_average_rounded} calories/kg")
    print(f"Percentage of the standard 2000 calories: {percentage_average_rounded}%")
else:
    print("One or both of the columns 'Q10 clean (Male kcal)' and 'Q27 Clean (Male Macro)' do not exist in the DataFrame. Please check the correct column names.")
```

```python
# Check if both columns exist in the DataFrame
if 'Q11 Clean (female kcal)' in df_cleaned.columns and 'Q26 Clean (Female (macro and kcal)' in df_cleaned.columns:
    # Combine the two series, ignoring NaN values from both
    combined_female_macros = pd.concat([df_cleaned['Q11 Clean (female kcal)'].dropna(), df_cleaned['Q26 Clean (Female (macro and kcal)'].dropna()])
    
    # Calculate the average of the combined data, excluding NaN values
    combined_female_macros_average = combined_female_macros.mean()
    
    # Calculate the rounded average (absolute value)
    absolute_female_macros_average_rounded = round(combined_female_macros_average, 1)
    
    # Calculate the relative value based on body weight (for a 60kg female)
    relative_female_macros_average_rounded = round(absolute_female_macros_average_rounded / 60, 1)  # per kg
    
    # Calculate the percentage increase based on the ending calories (1000 calories)
    percentage_female_macros_average_rounded = round((absolute_female_macros_average_rounded / 1000) * 100, 1)  # Convert to percentage
    
    # Print out the results
    print(f"Average daily total macronutrient increase (absolute): {absolute_female_macros_average_rounded} (assuming this is in grams or kcal as appropriate)")
    print(f"Relative daily total macronutrient increase per kg body weight (for a 60kg female): {relative_female_macros_average_rounded} (g or kcal/kg)")
    print(f"Percentage increase relative to 1000 calories: {percentage_female_macros_average_rounded}%")
else:
    print("One or both of the columns 'Q11 Clean (female kcal)' and 'Q26 Clean (Female (macro and kcal)' do not exist in the DataFrame. Please check the correct column names.")
```
